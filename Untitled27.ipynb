```python
def fact(number):
    temp=1
    while number >=1:
        temp=temp*number
        number -=1
    return temp
    
```

```python
l1=list(range(0,100))
```

```python
l2=list(range(100,200))
```

```python
def factorial(a):
    temp = 1
    while a >= 1:
        temp = temp*a
        a = a-1
    return temp

```

```python
l1=list(range(1,10))
l2=list(range(20,30))
l3=list(range(40,50))
for i in l1:
    if i%2==0:
        print(factorial(i))
for j in l2:
    if j%2!=0:
        print(factorial(j))
```
